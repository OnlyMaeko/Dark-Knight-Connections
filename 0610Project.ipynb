# Batman Begins

- Analyzing the social networks within the first movie in the trilogy "Batman Begins" 

### 1 : Gotham City Underworld 

```python
import networkx as nx

U = nx.Graph()
U.add_node("Carmine Falcone", label="Mob Boss")
U.add_node("Scarecrow", label="Criminal Psychologist")
U.add_node("Ra's al Ghul", label="Leader of the League of Shadows")

U.add_edge("Carmine Falcone", "Scarecrow", relationship="Partners")
U.add_edge("Scarecrow", "Ra's al Ghul", relationship="Mentor-Protégé")

nx.draw(U, with_labels=True, node_size=500, node_color='lightblue', font_size=10, font_color='black', font_weight='bold')



```

### 2 : Bruce Wayne's Social Network

```python
B = nx.Graph()
B.add_node("Bruce Wayne", label="Billionaire")
B.add_node("Alfred Pennyworth", label="Butler")
B.add_node("Lucius Fox", label="Wayne Enterprises CEO")
B.add_node("Rachel Dawes", label="Assistant DA")

B.add_edge("Bruce Wayne", "Alfred Pennyworth", relationship="Mentor-Father Figure")
B.add_edge("Bruce Wayne", "Lucius Fox", relationship="Business Associate")
B.add_edge("Bruce Wayne", "Rachel Dawes", relationship="Love Interest")

nx.draw(B, with_labels=True, node_size=500, node_color='lightblue', font_size=10, font_color='black', font_weight='bold')

```

### 3 : Police Department Network

```python
P = nx.Graph()
P.add_node("James Gordon", label="Detective")
P.add_node("GCPD Officers", label="Police Officers")

# Define connections between GCPD officers (e.g., partnerships, collaborations)
P.add_edge("James Gordon", "GCPD Officers", relationship="Professional Network")

nx.draw(P, with_labels=True, node_size=500, node_color='lightblue', font_size=10, font_color='black', font_weight='bold')
```

### 4 : Whole Movie Interactions

```python
import networkx as nx
import matplotlib.pyplot as plt

whole = nx.DiGraph()

characters = ["Bruce Wayne", "Alfred", "Rachel Dawes", "Gordon", "Lucius Fox", "Scarecrow", "Ra's Al Ghul"]
whole.add_nodes_from(characters)

interactions = [("Bruce Wayne", "Alfred"), ("Bruce Wayne", "Rachel Dawes"), ("Bruce Wayne", "Gordon"),
                ("Bruce Wayne", "Lucius Fox"), ("Bruce Wayne", "Scarecrow"), ("Rachel Dawes", "Gordon"),
                ("Rachel Dawes", "Lucius Fox"), ("Gordon", "Scarecrow"), ("Gordon", "Ra's Al Ghul")]

whole.add_edges_from(interactions)

pos = nx.spring_layout(whole, seed=56, k=0.8)

node_style = {
    "node_size": 1000,
    "node_color": "lightblue",
    "font_size": 12,
    "font_color": "black",
    "font_weight": "bold",
}

edge_style = {
    "width": 1.5,
    "arrowsize": 10,
}

nx.draw(whole, pos, with_labels=True, **node_style)
labels = {edge: edge[1] for edge in interactions}
edge_labels = {(u, v): v for u, v in interactions}
nx.draw_networkx_edge_labels(whole, pos, edge_labels=edge_labels, font_size=10, rotate=False, verticalalignment='center')

plt.title("Character Interactions")

plt.axis('off')

plt.show()
```
